```python
# import Pandas library and data

import pandas as pd

data = pd.read_csv("C:/Users/darla/Desktop/Titanic" + "/train_and_test2.csv")
data.head()
```

```python
# Separation of features(X) and Y target(Y)
X = data[['Age', 'Fare','Sex', 'sibsp', 'Parch', 'Pclass']]
Y = data['2urvived']

# Division into training and test sets
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=42)

print(X_train.shape, X_test.shape)
print(Y_train.shape, Y_test.shape)
print(X)
```

```python
# Training of the machine learning algorithm

from sklearn.linear_model import LogisticRegression # import LogisticRegression class

model = LogisticRegression() # Creates an instance of the logistic regression model
model.fit(X_train,Y_train) # Trains the model on training data

Y_pred = model.predict(X_test) # Use trained model to make predictions
Y_pred_proba = model.predict_proba(X_test) # Uses trained model to make probabilities
```

```python
# Evaluation of model accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test, Y_pred))
```

```python
# Model testing by a user

Age = float(input("Enter the passenger's age:"))
Fare = float(input("Enter the passenger's fare:"))
Sex = int(input("Enter the passenger's sex (0 for male, and 1 for female):"))
sibsp = int(input("Enter the passenger's sibsp (1 if sibsp, and 0 if not):"))
Parch = int(input("Enter the passenger's parch (1 if Parch, and 0 if not):"))
Pclass = int(input("Enter the passenger's Pclass (from 1 to 3):"))

Z = pd.DataFrame([{
    "Age": Age,
    "Fare": Fare,
    "Sex": Sex,
    "Sibsp": sibsp,
    "Parch": Parch,
    "Pclass": Pclass,
}])

Z_pred = model.predict(Z)
print(Z_pred)

Z_pred_proba = model.predict_proba(Z)
print(Z_pred_proba)

if Z_pred == [0]:
    print ("The passenger did not survive. His chances of survival was", int((Z_pred_proba[0][1]) * 100), "%")
else:
    print ("The passenger survived. His chances of survival was", int((Z_pred_proba[0][1]) * 100), "%")
```
